En este primer paso importamos la biblioteca de pandas

```python
import pandas as pd
```

```python
datos = pd.read_csv('https://raw.githubusercontent.com/cienciadedatos/datos-de-miercoles/master/datos/2019/2019-08-07/felicidad.csv')
```

Hacer la limpieza de datos es una buena práctica para evitar erros posteriores en los cálculos, por ende, se procede a reemplazar los valores nulos con 0.

```python
# limpieza de datos
datos_limpio = datos.fillna(0)
```

Para conocer cuál es el primer año de la base de datos, llamamos a la columna año donde buscamos el valor mínimo. En este caso el primer año es 2005. Adicional mediante este resultado también podemos ver los datos que contiene nuestra base.

```python
# Primer año
primer_anio = datos_limpio['anio'].min()  
datos_primer_anio_mundo = datos_limpio[datos_limpio['anio'] == primer_anio]
datos_primer_anio_mundo
```

De igual forma para encontrar el último año, de la columna año seleccionamos el valor máximo, en este caso sería el año 2018

```python
# Datos del último año
ultimo_anio = datos_limpio['anio'].max()
datos_ultimo_anio_mundo = datos_limpio[datos_limpio['anio'] == ultimo_anio]
datos_ultimo_anio_mundo
```

En los siguientes códigos lo que hacemos es calcular el promedio y la desviación estándar de la columna escalera_vida, misma que para el presente análisis representa la falicidad para cada uno de los países a nivel mundial. En estos casos específicos hacemos el cálculo para todos los países, es decir, cálculo a nivel mundial tanto para el primer año (2005) como para el último año 2018.

```python
# Promedio y desviación estándar primer año
mean_primer_anio_mundo = round(datos_primer_anio['escalera_vida'].mean(), 2)
sd_primer_anio_mundo = round(datos_primer_anio['escalera_vida'].std(), 2)
print(mean_primer_anio_mundo,sd_primer_anio_mundo)
```

```python
# Promedio y desviación estándar último año
mean_ultimo_anio_mudo = round(datos_ultimo_anio['escalera_vida'].mean(), 2)
sd_ultimo_anio_mundo = round(datos_ultimo_anio['escalera_vida'].std(), 2)
print(mean_ultimo_anio_mudo, sd_ultimo_anio_mundo)
```

Con la finalidad de comparar estos valores a nivel mundial con un país en específico, filtramos los datos para el primer año (2005) de Brasil (país seleccionado), mismo que nos da como resultado todos los valores registrados para Brasil solo del año 2005.

```python
# Datos Brasil primer año
datos_BRL_primer_anio = datos_limpio[(datos_limpio['pais'] == 'Brasil') & (datos_limpio['anio'] == 2005)]
datos_BRL_primer_anio
```

Repetimos los pasos anteriores, pero en este caso para Canadá.

```python
# Datos Canada primer año 
datos_CA_primer_anio = datos_limpio[(datos_limpio['pais'] == 'Canadá') & (datos_limpio['anio'] == 2005)]
datos_CA_primer_anio
```

Aquí nuevamente filtramos los datos, pero en este caso para el último año, es decir 2018. Esto tanto para Brasil como para Canadá.

```python
# Datos Brasil y último año
datos_BRL_ultimo_anio = datos_limpio[(datos_limpio['pais'] == 'Brasil') & (datos_limpio['anio'] == 2018)]
datos_BRL_ultimo_anio
```

```python
# Datos Canadá último año
datos_CA_ultimo_anio = datos_limpio[(datos_limpio['pais'] == 'Canadá') & (datos_limpio['anio'] == 2018)]
datos_CA_ultimo_anio
```

En este último paso se hizo una tabla resumen con la finalidad de tener una mejor visualización de los datos. Para ello se hizo un Dataframe a partir de una biblioteca con los datos calculados anteriormente.

```python
datos_resumen = {'Año': [2005, 2018],
        'Mundo mean': [6.45, 5.5],
        'Mundo sd': [0.92, 1.1],         
        'Brasil': [6.64, 6.20],
        'Canada': [7.42, 7.18]}

df_datos_resumen = pd.DataFrame(datos_resumen)
df_datos_resumen
```

Los resultados nos permiten aprecia que la felicidad a nivel mundial ha ido disminuyendo, datos que el promedio en el año 2005 era de 6.45 y en el 2018 baja a 5.5. De igual forma, se puede aprecia que tanto Brasil como Canadá superan el promedio mundial de felicidad en elaño 2005 y 2018, además, entre Brasil y Canadá, Canadá presenta se muestra como un país más faliz.

Respecto a la desviación estándar, vemos que para el año 2018 aumenta respecto al 2005, esto indica que los valores de la felicidad se vuelven mucho más dispersos a medida que van aumentado los años.
